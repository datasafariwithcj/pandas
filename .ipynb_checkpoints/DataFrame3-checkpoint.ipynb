```python
bond = pd.read_csv("jamesbond.csv")
bond.head()
```

```python
def rank_movie(row):
    year = row.loc["Year"]
    actor = row.loc["Actor"]
    budget = row.loc["Budget"]
    
    if year >= 1980 and year < 1990:
        return "Great 80's flick!"
    if actor == "Roger Moore":
        return "The best Bond ever!"
    if budget > 100:
        return "Expensive movie, fun"
    return "No comment"
    
bond.insert(loc = len(bond.columns), column = "My comments", value = bond.apply(rank_movie, axis = "columns"))
bond
```
