```python
import pandas as pd
```

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
fortune.head()
```

```python
fortune.pivot_table(index = "Sector", values = "Revenue", aggfunc = "sum")
```

```python
sectors = fortune.groupby(by = "Sector")
sectors
```

```python
len(sectors)
```

```python
sectors.size()
```

```python
sectors.first()
```

```python
sectors.last()
```

```python
sectors[["Revenue", "Employees"]].sum()
```

```python
sectors = fortune.groupby(by = ["Sector", "Industry"])
sectors["Revenue"].size()
```

```python
sectors["Revenue"].sum()
```

```python
sectors.agg({"Revenue" : "sum", "Employees" : "count"})
```
