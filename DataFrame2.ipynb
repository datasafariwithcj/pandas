```python
employees = pd.read_csv("employees.csv", parse_dates = ["Start Date"], date_format = "%m/%d/%Y") # Alternative aproach
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees
```

```python
employees[employees["Gender"] == "Male"]
```

```python
employees[employees["Team"]=="Finance"]
```

```python
employees[employees["Senior Management"]] # Already as bool series, so no comparison operator needed. 
```

```python
employees[employees["Salary"]>110000]
```

```python
employees[employees["Bonus %"] < 1.5]
```

```python
employees[employees["Start Date"] >= "1985-01-01"]
```

```python
employees[employees["Last Login Time"] < dt.time(12,0,0)]
```

```python
employees[(employees["Gender"] == "Female") & (employees["Team"] == "Marketing") & (employees["Salary"] > 100000)]
```

```python
employees[(employees["Senior Management"]) | (employees["Start Date"] < "1991-01-01")]
```

```python
employees[((employees["First Name"] == "Robert") & (employees["Team"] == "Client Services")) | (employees["Start Date"] > "2016-06-01")] 
```
